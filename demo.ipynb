```python
# 右边乘数 = 当前的行数
# 左边乘数 = 从1 - 当前的行数

# 行号初始值
row_num = 1

# 当行号小于9的时候，循环输出
while row_num <= 9:
    left = 1
    # 内部循环是控制 左边乘数
    while left <= row_num:
        print('%d*%d=%-2d  ' % (left, row_num, left*row_num), end='')
        # print(f'{left}*{row_num}={left*row_num}  ', end='')
        # 循环变量
        left += 1
    row_num += 1
    print('')
```
